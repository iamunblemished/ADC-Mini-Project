```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Function to modulate a signal using AM
def am_modulate(carrier_freq, modulation_index, signal, time):
    carrier_wave = np.sin(2 * np.pi * carrier_freq * time)
    modulated_wave = (1 + modulation_index * signal) * carrier_wave
    return modulated_wave

# Function to demodulate an AM signal
def am_demodulate(modulated_wave, carrier_freq, time):
    carrier_wave = np.sin(2 * np.pi * carrier_freq * time)
    demodulated_wave = modulated_wave / carrier_wave
    return demodulated_wave

# Function to plot the PSD spectrum
def plot_psd(signal, title):
    plt.figure()
    plt.psd(signal, NFFT=1024, Fs=2, scale_by_freq=True)
    plt.title(title)
    plt.show()

# Load a sample voice signal (you can replace this with an image loading mechanism)
sample_rate, voice_signal = wavfile.read("entry.wav")
time = np.arange(0, len(voice_signal) / sample_rate, 1 / sample_rate)

# Modulation Parameters
carrier_frequency = 1000  # Adjust this value as needed
modulation_indices = [0.5, 1, 1.5]

# Modulate and Demodulate for different modulation indices
for modulation_index in modulation_indices:
    # Modulate
    modulated_signal = am_modulate(carrier_frequency, modulation_index, voice_signal, time)

    # Demodulate
    demodulated_signal = am_demodulate(modulated_signal, carrier_frequency, time)

    # Plot original and demodulated signals
    plt.figure()
    plt.plot(time, voice_signal, label='Original Voice Signal')
    plt.plot(time, demodulated_signal, label=f'Demodulated Signal (m={modulation_index})')
    plt.title('AM Voice Transmission')
    plt.legend()
    plt.show()

    # Plot PSD Spectrum
    plot_psd(modulated_signal, f'PSD Spectrum (m={modulation_index})')
```
